```python
import random
import time

suits = ('Spades', 'Hearts', 'Diamonds', 'Clubs')
ranks = ('Ace', 'Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King')
values = {'Ace': 14, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5, 'Six': 6, 'Seven': 7, 
         'Eight': 8, 'Nine': 9, 'Ten':10, 'Jack': 11, 'Queen': 12, 'King': 13}
    
# Card Class
class Card():
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
    def __str__(self):
        return f'{self.rank} of {self.suit}'
        
class Deck():
    def __init__(self, owner = 'Dealer'):
        self.owner = owner
        self.deck = []
        for suit in suits:
            for rank in ranks:
                created_card = Card(suit, rank)
                self.deck.append(created_card)
    def shuffle_deck(self):
        random.shuffle(self.deck)
    def deal_one(self):
        return self.deck.pop()

# Player Class
class Player():
    def __init__(self,player_name = 'CPU'):
        self.name = player_name
        self.hand = []
    def deal_one(self):
        return self.hand.pop(random.randint(0,len(self.hand)-1))
    def collect_cards(self, cards):
        if type(cards) == type([]):
            self.hand.extend(cards)
        else:
            self.hand.append(cards)
    def __str__(self):
        return f'{self.name} now has {len(self.hand)} card(s)'
        
```

```python
new_deck = Deck('Longines')
mark = Player('Mark')
cpu = Player()

new_deck.shuffle_deck()
for i in range(26):
    mark.collect_cards(new_deck.deal_one())
    cpu.collect_cards(new_deck.deal_one())
# mark.collect_cards([new_deck.deal_one() for _ in range(int(len(new_deck.deck)/2))])
# cpu.collect_cards([new_deck.deal_one() for _ in range(len(new_deck.deck))])
# print(mark)
# print(cpu)
game = True
game_round = 0
war_card = 5
while game:
    
    cards_on_table = []
    at_war = True
    while at_war:
        if len(mark.hand) == 0 or len(cpu.hand) == 0:
            break
        at_war = False
        marks_card = mark.deal_one()
        cpus_card = cpu.deal_one()
        print(f'{mark.name} played {marks_card}')
        print(f'{cpu.name} played {cpus_card}')
        cards_on_table.append(marks_card)
        cards_on_table.append(cpus_card)
#         time.sleep(0.9)
        if marks_card.value == cpus_card.value:
            at_war = True
            print('We\'re at war!')
            if len(mark.hand) <= war_card or len(cpu.hand) <= war_card:
                for _ in range(len(mark.hand)):
#                     marks_card = mark.deal_one()
                    cards_on_table.append(mark.deal_one())
                for _ in range(len(cpu.hand)):
#                     cpus_card = cpu.deal_one()
                    cards_on_table.append(cpu.deal_one())
                break
            else:
                for _ in range(war_card):
                    marks_card = mark.deal_one()
                    cpus_card = cpu.deal_one()
                    cards_on_table.append(marks_card)
                    cards_on_table.append(cpus_card)
        elif marks_card.value > cpus_card.value:
            print(f'{mark.name} wins the round.')
            game_round += 1
            mark.collect_cards(cards_on_table)
            cards_on_table = []
        elif marks_card.value < cpus_card.value:
            print(f'{cpu.name} wins the round.')
            game_round += 1
            cpu.collect_cards(cards_on_table)
            cards_on_table = []
    if len(mark.hand) == 0 or len(cpu.hand) == 0:
        game = False
        print(f'After {game_round} round(s), {mark.name if len(cpu.hand) == 0 else cpu.name} wins the game!')
        break
    print(mark)
    print(cpu)
    print('--------------------------------')
#     time.sleep(1.5)
                                  
```
